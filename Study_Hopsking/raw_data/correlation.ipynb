**Heat Map**

```python
# Calcula la matriz de correlación
correlation_matrix = df.corr()

# Crea el mapa de calor
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Matriz de Correlación')
plt.show()
```

As we can see, the variables with the best correlation respect Real Power are:

1. TimeStamp
2. Month
3. Average_hour_month
4. Average_hour_day_ant

And the only variable with correlation respect EV_Power is Theres_sun
