**OBSERVACION: LA CARGA TOTAL DE TODOS LOS EDIFICIOS SE ESTA INTEGRANDO, TAL VEZ SURJA ERROR**

*Solución: Modelo con carga -1*

```python
data4=pd.read_csv("../UCSD-Microgrid-Database\Data Files\DemandCharge.csv")

data4['TotalCampusLoad_ant'] = data4['TotalCampusLoad'].shift(-1)
data4['OnCampusGeneration_ant'] = data4['OnCampusGeneration'].shift(-1)
data4['SDG&E Import_ant'] = data4['SDG&E Import'].shift(-1)
data4['AdjustedDemand_ant'] = data4['AdjustedDemand'].shift(-1)



data4 = data4[data4['AdjustedDemand_ant'].notna()]

data4
```

```python
del data4["AdjustedDemand"]
del data4["TotalCampusLoad"]
del data4["OnCampusGeneration"]
del data4["SDG&E Import"]
```

```python
data4["DateTime"]=pd.to_datetime(data4["DateTime"])

data_new2 = pd.merge(data1, data4, on='DateTime', how="outer")
data_new2 = data_new2.dropna(subset=['RealPower'])

data_new2
```

```python
data_new2.to_csv("data_minus1.csv")
```
