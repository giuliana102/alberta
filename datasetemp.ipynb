```python
# Package ID: knb-lter-cce.13.8 Cataloging System:https://pasta.edirepository.org.
# Data set title: Daily weather summaries (temperature, precipitation and wind) for station SAN DIEGO LINDBERGH FIELD, CA US (GHCND:USW00023188), obtained from NOAA National Climatic Data Center and re-published through CCE LTER data system for community use as regionally relevant data, 1939 - Oct 2019..
# Data set creator:    - NOAA National Climatic Data Center 
# Metadata Provider:    - California Current Ecosystem LTER 
# Contact:    - CCE LTER Information Manager California Current Ecosystem LTER  - ccelter-im@ucsd.edu
# Stylesheet v1.0 for metadata conversion into program: John H. Porter, Univ. Virginia, jporter@virginia.edu      
# 
# This program creates numbered PANDA dataframes named dt1,dt2,dt3...,
# one for each data table in the dataset. It also provides some basic
# summaries of their contents. NumPy and Pandas modules need to be installed
# for the program to run. 

import numpy as np
import pandas as pd 

infile1  ="https://pasta.lternet.edu/package/data/eml/knb-lter-cce/13/8/2e64236da92193f345c7b2281cf63e88".strip() 
infile1  = infile1.replace("https://","http://")
                 
dt1 =pd.read_csv(infile1 
          ,skiprows=1
            ,sep=","  
                ,quotechar='"' 
           , names=[
                    "Date",     
                    "Min_Temp",     
                    "Min_Temp_M_Flag",     
                    "Min_Temp_Q_Flag",     
                    "Min_Temp_S_Flag",     
                    "Max_Temp",     
                    "Max_Temp_M_Flag",     
                    "Max_Temp_Q_Flag",     
                    "Max_Temp_S_Flag",     
                    "Avg_Temp",     
                    "Precipitation",     
                    "Precipitation_M_Flag",     
                    "Precipitation_Q_Flag",     
                    "Precipitation_S_Flag",     
                    "Average_Wind_Speed",     
                    "Average_Wind_Speed_M_Flag",     
                    "Average_Wind_Speed_Q_Flag",     
                    "Average_Wind_Speed_S_Flag"    ]
# data type checking is commented out because it may cause data
# loads to fail if the data contains inconsistent values. Uncomment 
# the following lines to enable data type checking
         
#            ,dtype={ 
#             'Date':'str' , 
#             'Min_Temp':'float' ,  
#             'Min_Temp_M_Flag':'str' ,  
#             'Min_Temp_Q_Flag':'str' ,  
#             'Min_Temp_S_Flag':'str' , 
#             'Max_Temp':'float' ,  
#             'Max_Temp_M_Flag':'str' ,  
#             'Max_Temp_Q_Flag':'str' ,  
#             'Max_Temp_S_Flag':'str' , 
#             'Avg_Temp':'float' , 
#             'Precipitation':'float' ,  
#             'Precipitation_M_Flag':'str' ,  
#             'Precipitation_Q_Flag':'str' ,  
#             'Precipitation_S_Flag':'str' , 
#             'Average_Wind_Speed':'float' ,  
#             'Average_Wind_Speed_M_Flag':'str' ,  
#             'Average_Wind_Speed_Q_Flag':'str' ,  
#             'Average_Wind_Speed_S_Flag':'str'  
#        }
          ,parse_dates=[
                        'Date',
                ] 
            ,na_values={
                  'Date':[
                          'NULL',],
                  'Min_Temp':[
                          'NULL',
                          '-9999',
                          '9999',],
                  'Min_Temp_M_Flag':[
                          'NULL',],
                  'Min_Temp_Q_Flag':[
                          'NULL',],
                  'Min_Temp_S_Flag':[
                          'NULL',],
                  'Max_Temp':[
                          'NULL',
                          '-9999',
                          '9999',],
                  'Max_Temp_M_Flag':[
                          'NULL',],
                  'Max_Temp_Q_Flag':[
                          'NULL',],
                  'Max_Temp_S_Flag':[
                          'NULL',],
                  'Avg_Temp':[
                          'NULL',],
                  'Precipitation':[
                          'NULL',
                          '-9999',
                          '9999',],
                  'Precipitation_M_Flag':[
                          'NULL',],
                  'Precipitation_Q_Flag':[
                          'NULL',],
                  'Precipitation_S_Flag':[
                          'NULL',],
                  'Average_Wind_Speed':[
                          'NULL',
                          '-9999',
                          '9999',],
                  'Average_Wind_Speed_M_Flag':[
                          'NULL',],
                  'Average_Wind_Speed_Q_Flag':[
                          'NULL',],
                  'Average_Wind_Speed_S_Flag':[
                          'NULL',],} 
            
    )
# Coerce the data into the types specified in the metadata 
# Since date conversions are tricky, the coerced dates will go into a new column with _datetime appended
# This new column is added to the dataframe but does not show up in automated summaries below. 
dt1=dt1.assign(Date_datetime=pd.to_datetime(dt1.Date,errors='coerce')) 
dt1.Min_Temp=pd.to_numeric(dt1.Min_Temp,errors='coerce')  
dt1.Min_Temp_M_Flag=dt1.Min_Temp_M_Flag.astype('category')  
dt1.Min_Temp_Q_Flag=dt1.Min_Temp_Q_Flag.astype('category')  
dt1.Min_Temp_S_Flag=dt1.Min_Temp_S_Flag.astype('category') 
dt1.Max_Temp=pd.to_numeric(dt1.Max_Temp,errors='coerce')  
dt1.Max_Temp_M_Flag=dt1.Max_Temp_M_Flag.astype('category')  
dt1.Max_Temp_Q_Flag=dt1.Max_Temp_Q_Flag.astype('category')  
dt1.Max_Temp_S_Flag=dt1.Max_Temp_S_Flag.astype('category') 
dt1.Avg_Temp=pd.to_numeric(dt1.Avg_Temp,errors='coerce') 
dt1.Precipitation=pd.to_numeric(dt1.Precipitation,errors='coerce')  
dt1.Precipitation_M_Flag=dt1.Precipitation_M_Flag.astype('category')  
dt1.Precipitation_Q_Flag=dt1.Precipitation_Q_Flag.astype('category')  
dt1.Precipitation_S_Flag=dt1.Precipitation_S_Flag.astype('category') 
dt1.Average_Wind_Speed=pd.to_numeric(dt1.Average_Wind_Speed,errors='coerce')  
dt1.Average_Wind_Speed_M_Flag=dt1.Average_Wind_Speed_M_Flag.astype('category')  
dt1.Average_Wind_Speed_Q_Flag=dt1.Average_Wind_Speed_Q_Flag.astype('category')  
dt1.Average_Wind_Speed_S_Flag=dt1.Average_Wind_Speed_S_Flag.astype('category') 
      
print("Here is a description of the data frame dt1 and number of lines\n")
print(dt1.info())
print("--------------------\n\n")                
print("Here is a summary of numerical variables in the data frame dt1\n")
print(dt1.describe())
print("--------------------\n\n")                
                         
print("The analyses below are basic descriptions of the variables. After testing, they should be replaced.\n")                 

print(dt1.Date.describe())               
print("--------------------\n\n")
                    
print(dt1.Min_Temp.describe())               
print("--------------------\n\n")
                    
print(dt1.Min_Temp_M_Flag.describe())               
print("--------------------\n\n")
                    
print(dt1.Min_Temp_Q_Flag.describe())               
print("--------------------\n\n")
                    
print(dt1.Min_Temp_S_Flag.describe())               
print("--------------------\n\n")
                    
print(dt1.Max_Temp.describe())               
print("--------------------\n\n")
                    
print(dt1.Max_Temp_M_Flag.describe())               
print("--------------------\n\n")
                    
print(dt1.Max_Temp_Q_Flag.describe())               
print("--------------------\n\n")
                    
print(dt1.Max_Temp_S_Flag.describe())               
print("--------------------\n\n")
                    
print(dt1.Avg_Temp.describe())               
print("--------------------\n\n")
                    
print(dt1.Precipitation.describe())               
print("--------------------\n\n")
                    
print(dt1.Precipitation_M_Flag.describe())               
print("--------------------\n\n")
                    
print(dt1.Precipitation_Q_Flag.describe())               
print("--------------------\n\n")
                    
print(dt1.Precipitation_S_Flag.describe())               
print("--------------------\n\n")
                    
print(dt1.Average_Wind_Speed.describe())               
print("--------------------\n\n")
                    
print(dt1.Average_Wind_Speed_M_Flag.describe())               
print("--------------------\n\n")
                    
print(dt1.Average_Wind_Speed_Q_Flag.describe())               
print("--------------------\n\n")
                    
print(dt1.Average_Wind_Speed_S_Flag.describe())               
print("--------------------\n\n")
                    
                    
                




```
